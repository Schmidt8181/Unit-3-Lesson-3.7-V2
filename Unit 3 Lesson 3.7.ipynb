```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
mean = np.mean(pop1)
st_dev = np.std(pop1)
plt.axvline(x=mean, color='red')
plt.axvline(x=(mean + st_dev), color='red', linestyle='--')
plt.axvline(x=(mean - st_dev), color='red', linestyle='--')
mean = np.mean(pop2)
st_dev = np.std(pop2)
plt.axvline(x=mean, color='green')
plt.axvline(x=(mean + st_dev), color='green', linestyle='--')
plt.axvline(x=(mean - st_dev), color='green', linestyle='--')
plt.show()
print(pop1.mean())
print(pop2.mean())
print(pop1.std())
print(pop2.std())
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
plt.hist(sample1, alpha=0.5, label='Population 1') 
plt.hist(sample2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
mean = np.mean(sample1)
st_dev = np.std(sample1)
plt.axvline(x=mean, color='red')
plt.axvline(x=(mean + st_dev), color='red', linestyle='--')
plt.axvline(x=(mean - st_dev), color='red', linestyle='--')
mean = np.mean(sample2)
st_dev = np.std(sample2)
plt.axvline(x=mean, color='green')
plt.axvline(x=(mean + st_dev), color='green', linestyle='--')
plt.axvline(x=(mean - st_dev), color='green', linestyle='--')
plt.show()
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
```

```python
# After changing the sample size from 10000 data points to 100 data points
# the mean and mode moved away from each other.
# And the standard deviation grew, although only a little
```
